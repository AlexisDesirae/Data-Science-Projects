**Predicting Passanger Airline Satisifaction**

My dataset consists of data collected from an airline satisfaction survey. Data such as gender, age, on-board service, cleanliness, departure delays, etc. are captured in this survey.

I will use this dataset to predict airline satisfaction of different passangers who flew with the airline based on their experience.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns; sns.set()
```

Next, I will load my dataset "Airline Customer Satisfaction." I will create a varaiable out of the dataset and name it 'df'

I will use the Head() Method to return the top five rows in my dataset. This will show me each column's name

```python
df = pd.read_csv ('/content/Airline Customer Satisfaction.csv')
df.head()
```

I will use Info() Method to find out how many entries my dataset has, how many columns, and for insight on each column

```python
df.info
```

I use LabelEncoder to convert the class and satisfaction text into numbers because they are ordinal values and need to be ranked accordingly 

Business class is higher than Econ Plus which is higher than Econ

Satisfied is higher than neutral or dissatisfied

```python
from sklearn.preprocessing import LabelEncoder
le_Class = LabelEncoder()
le_satisfaction = LabelEncoder()
```

 I will create a new column with the new values and name them 'Class_n' and 'satisfaction_n'

```python
df['Class_n'] = le_Class.fit_transform(df['Class'])
df['satisfaction_n'] = le_satisfaction.fit_transform(df['satisfaction'])
```

Here I will create a new table with the new class and satisfaction columns and name the new table "satisfaction"

```python
satisfaction = df
satisfaction
```

I use get_dummies to differentiate between male and female and personal trip and business trip in a way that does not make it one category seem "better" than another. 

The variables are equal so to treat them as such I use get_dummies which will assign one category the number 0 and the other category the number 1 for both the Gender column and the Type of Travel column

```python
dummies = pd.get_dummies(data=satisfaction, columns=['Gender', 'Type of Travel'])
dummies
```

Here I will drop all NaN values

```python
dummies.dropna(axis='columns', how='all')
```

Identify our independent or target variable (y) and our dependent variable(s) X.

To identify our X we must create a table that includes only the dependent variables that we want to test against our independent variable. I used the Drop() Method to drop all of the unnessesary columns and to drop the 'satisfaction' column because that will serve as my y variable. 

```python
X = dummies.drop(['Unnamed: 0', 'id', 'Customer Type','Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient','Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Cleanliness','satisfaction', 'Gender_Male', 'Type of Travel_Personal Travel', 'Class', 'satisfaction_n','Arrival Delay in Minutes', 'Ease of Online booking'], axis='columns')
X
```

I created a variable and named it 'y' and it includes the 'satisfaction' column which is my independent variable.

```python
y = dummies['satisfaction_n']
y
```

Now we will use train test split to evaluate the performance of each model.

I'll test LinearRegression first so I import LinearRegression from sklearn


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Then I fit the training data with model.fit


```python
model.fit(X_train, y_train)
```

Now I will score my data or test it for accuracy

```python
model.score(X_test, y_test)
```

 My Linear Regression model is only 31% accurate which is very low. So I will repeat the same process for Random Forest

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X, y)
```

```python
model.score(X,y)
```

Random Forest provided a much better accuracy of 87%
